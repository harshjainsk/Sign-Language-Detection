```python
import cv2
import os
import time
import uuid
```

```python
IMAGES_PATH = 'Tensorflow\workspace\images\collectedimages'
```

# Creating image labels and number of images to capture

```python
label = ['ayee_yo', 'yes', 'no', 'shocked', 'stop']
number_of_images = 20
```

# Creating our own dataset

```python
# !mkdir {'Tensorflow\\workspace\\images\\collectedimages\\' +label[0]}

cap = cv2.VideoCapture(0)
print(f"Collecting images for {label[1]}")

# time.sleep(5)
if cap.isOpened():
    for image_number in range(number_of_images):
        ret, frame = cap.read()
        imagename = os.path.join(IMAGES_PATH, label[1],label[1] +'_' + str(image_number) + f'_{str(uuid.uuid1())}.jpg')
        
        if ret:
#             img_name = os.path.join(IMAGES_PATH, label[0], str(uuid.uuid4()) + '.jpg')

            print(imagename)
            cv2.imwrite(imagename, frame)
            cv2.imshow('frame', frame)

            time.sleep(3)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
cap.release()
cap.destroyAllWindows()
```
